```python
import pandas as pd

canon = pd.read_csv('../data/char_freq_canon.csv')
ff = pd.read_csv('../data/char_freq_ff_mapped.csv')
genders = pd.read_csv('../data/genders.csv')
```

```python
merged = pd.merge(canon, ff, how='outer')
merged.fillna(0, inplace=True)
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', 3):
    print(merged)
```

```python
merged_genders = pd.merge(merged, genders)
merged_genders
```

```python
merged_genders.to_csv('../data/d3/char_frequencies_canon_ff.csv', index=False)
```
