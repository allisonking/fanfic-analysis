```python
import pandas as pd

file = '../data/char_freq_canon.txt'

with open(file) as f:
    content = f.readlines()

content = [line.strip() for line in content]
content[:10]
```

```python
char_freqs = []
for line in content:
    if line == '':
        continue
    else:
        split = line.split(' ', 1)
        char_freqs.append(split)
```

```python
char_freqs
```

```python
char_freqs_df = pd.DataFrame(char_freqs, columns=['count','name'])
```

```python
char_freqs_df
```

```python
char_freqs_df_reorder = char_freqs_df[['name','count']]
```

```python
df = char_freqs_df_reorder.set_index(['name'])
```

```python
df_asint = df.astype(int)
```

```python
percents = df_asint/df_asint.sum()
```

```python
percents.columns=['canon_percentage']
```

```python
percents
```

```python
percents.to_csv('../data/char_freq_canon.csv')
```
